```python
from databricks.sdk import WorkspaceClient
from databricks.labs.dqx.profiler.profiler import DQProfiler

w = WorkspaceClient()
profiler = DQProfiler(workspace_client=w, spark=spark)

print("DQX Profiler initialized!")
```

```python
chicago = spark.table("workspace.default.bronze_chicago_inspections")
chicago_profile = profiler.profile(chicago)

print("=== CHICAGO DQX PROFILE ===")
for p in chicago_profile:
    print(p)
```

```python
dallas = spark.table("workspace.default.bronze_dallas_inspections")
dallas_profile = profiler.profile(dallas)

print("=== DALLAS DQX PROFILE (Core columns) ===")
for p in dallas_profile:
    col_name = str(p.get("col_name", p.get("column", "")))
    if not col_name.startswith("violation_"):
        print(p)
```

```python
print("=== DALLAS DQX PROFILE (Violation columns) ===")
for p in dallas_profile:
    col_name = p.get("col_name", p.get("column", ""))
    if col_name.startswith("violation_"):
        print(p)
```
